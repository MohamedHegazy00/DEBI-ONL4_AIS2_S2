```python
# TASK 1


email="Amit_ml@gmail.edu"
if email.count("@") != 1:
    print("inValid Email")
else:
    username, domain = email.split("@")
print("Username:", username)
print("Domain:", domain)  
if domain.endswith(".edu"):
    print("Educational Domain")
elif domain.endswith(".com"):
    print("Commercial Domain")
else:
    print("Other Domain")    

```

```python
#TASK 2


message = "###!!@mocleW EPGTQ!!!6789"

start = message.index("@") + 1
end = message.find("!", start)   

core_part = message[start:end]

first_word, second_word = core_part.split()

decoded_first = first_word[::-1]
decoded_second = second_word

decoded_message = decoded_first + " " + decoded_second

print(decoded_message)

```

```python
# TASK 3


message = "&&&**$gnirtS PLIO!!@1234"


start = message.index("$") + 1
end = message.index("!")
core_part = message[start:end]        


first_word, second_word = core_part.split()


decoded_first = first_word[::-1]     


decoded_second = (
    second_word.replace("I", "E")
               .replace("O", "U")
)

decoded_message = decoded_first + " " + decoded_second

print(decoded_message)
```

```python
# TASK 4

message = "##$$$@!yalpstcejorp EPUVT****9887"


start = message.index("!") + 1
end = message.index("*")
core_part = message[start:end]          


first_word, second_word = core_part.split()


decoded_first = first_word[::-1]       


decoded_second = (
    second_word.replace("E", "A")
               .replace("U", "O")
)

decoded_message = decoded_first + " " + decoded_second

print(decoded_message)
```
