```python
import os
import random

def thanos_project(folder_path):
    # 1. Open the whole folder
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    # 2. Create number of files inside it (creating 10 files as example)
    for i in range(1, 11):
        file_name = f"file_{i}.txt"
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'w') as f:
            f.write(f"This is file number {i}")
    
    # 3. Check the len of files in that folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    print(f"Total files before snap: {len(files)}")
    
    # 4. Delete random half of this folder
    half = len(files) // 2
    files_to_delete = random.sample(files, half)
    
    for file_name in files_to_delete:
        file_path = os.path.join(folder_path, file_name)
        os.remove(file_path)
    
    # 5. Check the result
    remaining_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    print(f"Total files after snap: {len(remaining_files)}")
    print("Perfectly balanced, as all things should be.")

# Run the function
thanos_project(r"C:\github depi\DEBI-ONL4_AIS2_S2\python\session_2\new_directory")
```
