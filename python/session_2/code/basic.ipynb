```python
def user_info(name, age, city):
    """
    this function prints user information.
    
    :param name: this is the name of the user
    :param age: this is the age of the user
    :param city: this is the city where the user lives
    """
    return f" My Name is: {name}, Age: {age}, City: {city}"
```

```python
def reverse_string(s):
    split_s = s.split()
    new_list = split_s[::-1]
    result= " ".join(new_list)   
    return result
```

```python
def password_generator(length):

    """ This Function Generates a random password of given length entered by user
    ."""

    import random
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ''.join(random.choice(characters) for i in range(length))
    return f"Generated password: {password}"
```
